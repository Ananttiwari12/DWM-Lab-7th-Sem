handle missing value through mode

```python
df_filled_freq=df.copy()
```

```python
for col in df_filled_freq.columns:
  if df_filled_freq[col].dtype in [np.float64, np.int64]:
    mode_value = df_filled_freq[col].mode()[0]
    df_filled_freq[col].fillna(mode_value, inplace=True)
```

#Dataset: Bengaluru_House_data.csv

```python
df2= pd.read_csv('Bengaluru_House_Data.csv')
```

```python
df2
```

```python
df_mean=df2.copy()
df_median=df2.copy()
df_mode=df2.copy()

numerical_columns = df2.select_dtypes(include=[np.number]).columns

for col in numerical_columns:
    mean_value = df_mean[col].mean()
    median_value = df_median[col].median()
    most_frequent_value = df_mode[col].mode()[0]

    df_mean[col].fillna(mean_value, inplace=True)
    df_median[col].fillna(median_value, inplace=True)
    df_mode[col].fillna(most_frequent_value, inplace=True)
```

```python
# Example of handling missing values for categorical columns
categorical_columns_mode = df_mode.select_dtypes(include=[object]).columns
categorical_columns_mean = df_mean.select_dtypes(include=[object]).columns
categorical_columns_medain = df_median.select_dtypes(include=[object]).columns

for col in categorical_columns_mode:
    most_frequent_value = df_mode[col].mode()[0]
    df_mode[col].fillna(most_frequent_value, inplace=True)

for col in categorical_columns_mean:
    most_frequent_value = df_mean[col].mode()[0]
    df_mean[col].fillna(most_frequent_value, inplace=True)

for col in categorical_columns_medain:
    most_frequent_value = df_median[col].mode()[0]
    df_median[col].fillna(most_frequent_value, inplace=True)
```
