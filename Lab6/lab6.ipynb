```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LogisticRegression
```

```python
X = np.array([[1, 2, 3, 4, 5, 6, 7, 8]]).T  # Feature matrix with shape (8, 1)
y = np.array([0, 0, 0, 0, 1, 1, 1, 1])      # Target array

# Create and fit the logistic regression model
model = LogisticRegression()
model.fit(X, y)
```

```python
X_test = np.linspace(1, 8, 300).reshape(-1, 1)

# Predict probabilities for each point in X_test
y_prob = model.predict_proba(X_test)[:, 1]

# Plotting
plt.figure(figsize=(8, 5))
plt.scatter(X, y, color='red', label='Data Points')  # Original data points
plt.plot(X_test, y_prob, color='blue', label='Logistic Regression Curve')  # Logistic curve
plt.xlabel('X')
plt.ylabel('Probability of y=1')
plt.title('Logistic Regression Curve')
plt.legend()
plt.grid(True)
plt.show()
```
